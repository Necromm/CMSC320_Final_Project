```python
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {}
    counter = 0
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary
        # and add it to data
        for rows in csvReader:
             
            # Assuming a column named 'No' to
            # be the primary key
            key = counter
            counter += 1
            data[key] = rows
 
    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
```

```python
def cleanCSV(filePath):
    f = open(filePath)
    text = f.read()
    f.close()
    text = text.replace('TFT6_', '')
    text = text.replace('Set6_', '')
    text = text.replace('TFT_Item_', '')
    text = text.replace('Augment_', '')
    text = text.replace(', 10006', '')
    # print(text)
    f = open(filePath,'w')
    f.write(text)
    f.close()

    dict_from_csv = {}
    with open(filePath, mode='r') as inp:
        reader = csv.reader(inp)
        dict_from_csv = {rows[0]:rows[1] for rows in reader}
```

```python
def cleanUnits(units):
  print('hi')
  for unit in units:
    keys = list(unit.keys())
    if 'itemNames' in keys:
      unit.pop('itemNames')
    if 'names' in keys:
      unit.pop('name')
```

```python
with open(jsonfile) as json_file:
    data = json.load(json_file)
```
